```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Valid temperature for Unruh Accretion

```python
# all in SI unit
kB = 8.31 / 6.02e23
h = 6.626e-34 
c = 3e8
G = 6.673e-11
mb = 1.67e-27

MPBH = np.logspace(13, 20)
def Tmax(M):
    return h ** 2 * c ** 4 / (8 * G ** 2 * M ** 2 * mb * kB)

plt.figure(figsize=(5, 4))
plt.loglog(MPBH, Tmax(MPBH))
plt.xlabel("PBH Mass")
plt.ylabel("Maximum Temperature")
plt.savefig("./img/Tmax.pgf")
```
