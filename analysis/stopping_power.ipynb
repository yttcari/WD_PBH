```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cross_section = 154 * 1e-3 * 1e-24 # cm^-2
rho = 1 * (9.79e5 / 0.5) # g/cc
mp = 1.67262192e-27 * 1000 # g
n_ion = rho / mp
R_stop = 1e-6 # cm
c = 3e10 # cm/s
me = 9.10938e-28 #g
barn = 1e-24 # cm^2
```

```python
def inel(nion, E, sigma=100*1e-3*barn):
    return nion * sigma * E

def elec(alpha, nion, sigma, E):
    return alpha * nion * sigma * E * np.log(E/me)

rho = 1e6 # g/cc
nion = rho / (6 * mp)
```

```python
plt.figure(figsize=(5, 4))
E = np.logspace(-1, 1)
plt.plot(E, inel(nion, E), label="Inelastic scattering")
plt.plot(E, inel(nion, sigma=1*1e-3*barn, E=E), label="Photonuclear")
plt.plot(E, elec(alpha=1/100, nion=nion, sigma=1*1e-3*barn, E=E), label="Electronuclear")

plt.xscale('log')
plt.legend()
plt.yscale('log')
plt.xlabel("Energy (GeV)")
plt.ylabel(r"$dE/dr$")
plt.savefig("./img/stoppingpower.pgf")
```
